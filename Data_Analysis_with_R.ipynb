## Simple Data Analysis with R

```r
library(tidyverse)
```

```r
View(mpg)
```

```r
glimpse(mpg)
```

```r
unique(mpg$cty)
```

```r
filter(mpg, cty>=23)
```

```r
dim(filter(mpg, cty>=23))
```

```r
mpg_ford <- filter(mpg,manufacturer == "ford")
head(mpg_ford)
```

```r
mpg_metric <- mutate(mpg, cty_metric = 0.425144 * cty) #1 Mile per gallon = 0.425143707 kilometers per liter
head(mpg_metric)
```

```r
mpg_metric <- mpg %>%
  mutate(cty_metric = 0.425144 * cty) 
  #It gives the same output, but this code helps make it more readable by chaining the operations(%>%).
head(mpg_metric) 
```

```r
mpg %>%
  group_by(class) %>%
  summarise(mean_cty = mean(cty),
            median_cty = median(cty),
            variance_cty = var(cty),
            standard_deviation_cty = sd(cty))
```

## Data Wiz With ggplot2

```r
ggplot(mpg, aes(x=cty)) +
  geom_histogram(binwidth = 1) +
  labs(x= "City Mileage")
```

```r
ggplot(mpg, aes(x=cty)) +
  geom_freqpoly(binwidth=1) +
  labs(x= "City Mileage")
```

```r
ggplot(mpg,aes(x=cty,
               y = hwy))+
  geom_point()+
  geom_smooth(method = "lm", formula = y ~ x)
```

```r
ggplot(mpg,aes(x=cty,
               y = hwy,
               color = class))+
  geom_point() +
  scale_color_brewer(palette = "Dark2")
```
