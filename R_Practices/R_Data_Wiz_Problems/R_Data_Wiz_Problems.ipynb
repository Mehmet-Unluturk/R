###  Problem 1
#### The teams supported by 25 people are given below. Create a frequency chart of the data and draw the appropriate graph.

```r
library(readxl)
```

```r
df <- read_excel("/data/notebook_files/Data.xlsx")
```

```r
head(df)
```

```r
frequency <- table(df$Team)
print(frequency)
```

```r
pie(frequency)
```

```r
pie(frequency, col=c("black", "yellow", "red", "blue"))
```

```r
percentage<-round(frequency/sum(frequency)*100)
inf<-paste(names(frequency),"%",percentage,sep=" ")
pie(frequency, labels=inf, main="Pie chart of teams", col=c("black", "yellow", "red", "blue"))
```

### Problem 2
#### WPI and CPI values in Turkey between 2003-2004 are given below. Draw the appropriate graph for the data.

```r
df2 <- read_excel("/data/notebook_files/Data2.xlsx")
View(df2)
```

```r
plot(df2$WPI, type="o", main="WPI change by months")
```

```r
plot(df2$CPI, type="o" , main="CPI change by months")
```

```r
plot(df2$WPI, col = "blue", lwd = 3, type = "l", xlab = "WPI", ylab = "CPI", main = "WPI - CPI Relationship")
lines(df2$CPI, col = "red", lwd = 3)
legend("topright", legend = c("WPI", "CPI"), col = c("blue", "red"), lwd = 3)
```

### Problem 3
#### Provincial/district center and town/village population percentage (%) values are given according to age group and gender for 2019. Draw the population pyramid.

```r
df3 <- read_excel("/data/notebook_files/Data3.xlsx")
View(df3)
```

```r
library(tidyverse)
```

```r
glimpse(df3)
```

```r
df3$Male <- as.integer(df3$Male)
df3$Female <- as.integer(df3$Female)
names(df3)[names(df3) == "Age Group"] <- "Age_Group"
```

```r
glimpse(df3)
```

```r
names(df3$Male)<-c(df3$Age_Group)
```

```r
pyramids(Left=df3$Male, Llab="Erkek", Right=df3$Female, Rlab="Kadin",Clab =df3$Age_Group)
```
